# **Project Objectives | Problem Statements**

## 1. Shipping Method Efficiency (Shipping Method vs Weight)
**Problem**: Identify which shipping methods handle the most weight across different countries. This can highlight which shipping methods are being overutilized or underutilized, allowing for optimization in logistics.  
**Solution**: Analyze whether heavier shipments are being transported via more cost-effective or efficient shipping methods.

## 2. Shipping Cost Optimization (Shipping Method vs Weight)
**Problem**: For countries where certain shipping methods carry a disproportionate amount of weight, the cost-effectiveness of the shipping method needs to be evaluated. Heavier shipments might require renegotiation of shipping contracts to lower costs.  
**Solution**: Explore renegotiation of shipping rates or the adoption of alternative shipping methods for high-weight goods.

## 3. Seasonal Trends in Sales (Date vs Value)
**Problem**: Detect seasonal peaks and troughs in sales value. Companies can experience fluctuating demand during holidays, special events, or specific seasons.  
**Solution**: Plan inventory and marketing strategies around high-value periods to maximize sales.

## 4. Demand Fluctuations Over Time (Date vs Quantity)
**Problem**: Analyze fluctuations in quantity sold over time to better understand how demand changes. These patterns can be caused by changes in customer preferences, external factors, or promotional efforts.  
**Solution**: Implement demand forecasting models to anticipate future sales trends and optimize inventory management.

## 5. Customer Preferences (Date vs Value)
**Problem**: Examine how customer preferences change over time based on value trends. Are certain products or services gaining or losing value at specific times?  
**Solution**: Adjust product offerings or pricing strategies based on changing customer preferences.

## 6. Inefficiency in Shipping Methods (Shipping Method vs Payment Terms)
**Problem**: Discover inefficiencies in how shipping methods correlate with payment terms. Are certain payment terms favoring shipping methods that are more costly or slower?  
**Solution**: Align more favorable payment terms with more efficient or cost-effective shipping methods to streamline the supply chain.

## 7. Supplier Payment Term Optimization (Payment Terms vs Value)
**Problem**: Identify suppliers offering unfavorable payment terms for high-value transactions. Shorter payment terms can strain cash flow.  
**Solution**: Focus on renegotiating payment terms with suppliers that have a significant share of high-value transactions to improve working capital and cash flow.

## 8. Underperforming Shipping Methods (Shipping Method vs Weight)
**Problem**: Detect underperforming shipping methods that carry low weight but incur high costs, leading to inefficiency.  
**Solution**: Replace or optimize underutilized shipping methods to ensure each method is being used effectively.

## 9. Unfavorable Payment Terms in Key Markets (Shipping Method vs Payment Terms)
**Problem**: Discover key markets or suppliers offering unfavorable payment terms despite a large volume of transactions. This can negatively impact cash flow and profitability.  
**Solution**: Prioritize suppliers in these key markets for renegotiating more favorable payment terms, such as extended payment periods.

## 10. Supplier Relationship Risk (Payment Terms vs Value)
**Problem**: Identify suppliers with high-value transactions that have short or risky payment terms, posing a threat to cash flow.  
**Solution**: Create a risk management plan to mitigate reliance on these suppliers and seek alternative partners or better terms.

## 11. Product Lifecycle Analysis (Date vs Quantity)
**Problem**: Evaluate the lifecycle of products by analyzing their quantity sold over time. Certain products may exhibit declining demand over time, indicating that they are nearing the end of their lifecycle.  
**Solution**: Prepare to phase out declining products and introduce new products to the market.

## 12. Sales Forecasting (Date vs Value)
**Problem**: Forecast future sales value based on historical patterns identified in the date vs value graph.  
**Solution**: Use past trends to predict sales and revenue growth or contraction, allowing for more precise planning and budgeting.

## 13. Payment Term Dependency on Shipping Method (Shipping Method vs Payment Terms)
**Problem**: Identify a dependency between certain shipping methods and specific payment terms. Are customers or suppliers favoring certain shipping methods based on the flexibility of payment terms?  
**Solution**: Adjust payment terms to be more favorable for more efficient shipping methods to reduce operational costs.

## 14. High-Volume, Low-Value Product Identification (Date vs Quantity & Shipping Method vs Weight)
**Problem**: Identify products or shipping methods that are high in volume but low in value, leading to inefficient resource allocation.  
**Solution**: Adjust product pricing or supply chain processes to ensure that high-volume transactions are more profitable.

```python
sample_df = df.sample(n = 3001, random_state = 55002)
# Convert 'Date' column to datetime format
sample_df['Date'] = pd.to_datetime(sample_df['Date'], errors='coerce')

# Drop any rows with invalid dates (NaT values)
sample_df = sample_df.dropna(subset=['Date'])

# Grouping sales by month to find seasonal trends
sample_df['Month'] = sample_df['Date'].dt.month
seasonal_sales = sample_df.groupby('Month')['Value'].mean()

print(f"Seasonal Sales Mean Values: \n{seasonal_sales}")
```

```python
# Splitting data into peak (e.g., June-August) and off-peak (e.g., December-February)
peak_sales = sample_df[(sample_df['Month'] >= 6) & (sample_df['Month'] <= 8)]['Quantity']
offpeak_sales = sample_df[(sample_df['Month'] >= 12) | (sample_df['Month'] <= 2)]['Quantity']

# Perform a t-test
t_stat, p_value = stats.ttest_ind(peak_sales, offpeak_sales)
print(f"T-test for Peak vs Off-Peak Sales: T-Statistic={t_stat}, P-value={p_value}")
```

```python
# Crosstab of product categories and sales periods
crosstab = pd.crosstab(sample_df['Category'], sample_df['Date'].dt.year)

# Perform chi-square test
chi2_stat, p_val, dof, ex = stats.chi2_contingency(crosstab)

print(f"Chi-Square test results: Chi2 Stat={chi2_stat}, P-value={p_val}")
```

```python
# Create a contingency table between shipping methods and payment terms
contingency_table = pd.crosstab(sample_df['Shipping_Method'], sample_df['Payment_Terms'])

# Perform chi-square test
chi2_stat, p_val, dof, ex = stats.chi2_contingency(contingency_table)

print(f"Chi-Square test for Shipping Method vs Payment Terms: Chi2 Stat={chi2_stat}, P-value={p_val}")
```

```python
# Calculate the 95th percentile for Value and 5th percentile for Quantity
high_value_threshold = sample_df['Value'].quantile(0.95)
low_quantity_threshold = sample_df['Quantity'].quantile(0.05)

# Filter transactions that have a Value greater than or equal to the 75th percentile
high_value_transactions = sample_df[sample_df['Value'] >= high_value_threshold]

# Filter transactions that have a Quantity less than or equal to the 25th percentile
low_quantity_transactions = sample_df[sample_df['Quantity'] <= low_quantity_threshold]

# Find the intersection of products that are in both filtered datasets (high value and low quantity)
intersection_products = pd.merge(high_value_transactions, low_quantity_transactions, on='Product')

# Display the products that are in both conditions
print(intersection_products)
```

```python
# Splitting the data into specific payment terms
cod_terms = sample_df[sample_df['Payment_Terms'] == 'Cash on Delivery']['Value'].dropna()
prepaid_terms = sample_df[sample_df['Payment_Terms'] == 'Prepaid']['Value'].dropna()
net_60_terms = sample_df[sample_df['Payment_Terms'] == 'Net 60']['Value'].dropna()
net_30_terms = sample_df[sample_df['Payment_Terms'] == 'Net 30']['Value'].dropna()

# Perform t-tests between different payment terms
t_stat_cod_vs_prepaid, p_val_cod_vs_prepaid = stats.ttest_ind(cod_terms, prepaid_terms)
t_stat_cod_vs_net60, p_val_cod_vs_net60 = stats.ttest_ind(cod_terms, net_60_terms)
t_stat_cod_vs_net30, p_val_cod_vs_net30 = stats.ttest_ind(cod_terms, net_30_terms)

# Print T-test results for each comparison
print(f"T-test Cash on Delivery vs Prepaid: T-Statistic={t_stat_cod_vs_prepaid}, P-value={p_val_cod_vs_prepaid}")
print(f"T-test Cash on Delivery vs Net 60: T-Statistic={t_stat_cod_vs_net60}, P-value={p_val_cod_vs_net60}")
print(f"T-test Cash on Delivery vs Net 30: T-Statistic={t_stat_cod_vs_net30}, P-value={p_val_cod_vs_net30}")
```
